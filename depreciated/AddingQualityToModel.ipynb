```python
df= df.reset_index()
```

```python
df["euDate"]=df["date"].dt.tz_localize("America/New_York").dt.tz_convert("Europe/Berlin")
```

```python
def resampleOHLCV(df,interval,how={"o":"first","h":"max","l":"min","c":"last"}):
    """Need to forwardfill new created columns in a concatenaed big df"""
    import re
    shift = timedelta(hours=1)
    i = int(re.match(r'\d+', interval).group())
    if interval.endswith("H"): shift = timedelta(hours=i)
    elif interval.endswith("D"): shift = timedelta(days=i)
    elif interval.endswith("min"): shift = timedelta(minutes=i)
    return df.resample(interval).agg(how)#.add_suffix(interval).shift(i, freq=shift)
```

```python
df=df.set_index("euDate")
```

```python
daily=resampleOHLCV(df,"1D",how={"o":"first","h":"max","l":"min","c":"last","date":"first"})
```

```python
daily=daily.dropna(subset=["c"])
```

```python
daily["body"] = np.sign(daily["c"] - daily["o"])
base = daily.index[0]

for i,row in daily.iterrows():
    if i<=base: continue
    daily.loc[i,"mom"] = int(1) if (row["body"]==daily.loc[base,"body"]) & (row["body"]>0) else int(-1) if (row["body"]==daily.loc[base,"body"]) & (row["body"]<0) else int(0)
    base = i
```

```python
daily=daily.reset_index().set_index("date")
```

```python
daily["mom"] = daily.mom.shift(1)
```

```python
daily=daily.dropna()
```

```python
df=df.reset_index().set_index("date")
```

```python
df.loc[df.index.intersection(daily.index),"mom"] = daily["mom"]
```

### Mean Decrease Accuracy
MDA feature importance with sample
weights, with purged k-fold CV, and with scoring by negative log-loss
or accuracy. It measures MDA importance as a function of the
improvement (from permutating to not permutating the feature),
relative to the maximum possible score (negative log-loss of 0, or
accuracy of 1). Note that, in some cases, the improvement may be
**negative**, meaning that the feature is actually **detrimental** to the
forecasting power of the ML algorithm.


```python
mda=mean_decrease_accuracy(clf2,X.loc[idx],y.loc[idx],cv_gen,sample_weight_train=w.loc[idx],scoring=log_loss)
plot_feature_importance(mda,
oob_score=clf2.oob_score_,
oos_score=oos.mean(),
)
```
